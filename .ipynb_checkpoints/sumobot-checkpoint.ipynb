```python
import turtle as t
from math import sqrt

t.clearscreen()
t.pu()
t.setpos(-400,0)
t.right(90)
t.pd()
t.circle(400)

t.pu()
t.goto(0,0)
t.pd()
t.dot('red')

class Sumobot():

    def __init__(self):

        self.done = False
        self.reward = 0
        self.hit, self.miss = 0, 0

        # Setup Background

        self.win = t.Screen()
        self.win.title('Sumobot')
        self.win.setup(width=1000, height=1000)
        self.win.tracer(0)

        # Sumobot

        self.sumobot = t.Turtle()
        self.sumobot.speed(0)
        self.sumobot.shape('square')
        self.sumobot.shapesize(stretch_wid=2, stretch_len=3)
        self.sumobot.color('blue')
        self.sumobot.penup()
        self.sumobot.goto(0, -380)
        self.sumobot.dx = 0
        self.sumobot.dy = 0

#         # Enemy

#         self.enemy = t.Turtle()
#         self.enemy.speed(0)
#         self.enemy.shape('square')
#         self.enemy.color('red')
#         self.enemy.penup()
#         self.enemy.shapesize(stretch_wid=4, stretch_len=4)
#         self.enemy.goto(0, 340)
#         self.enemy.dx = 3
#         self.enemy.dy = -3

        # Score

        self.score = t.Turtle()
        self.score.speed(0)
        self.score.color('black')
        self.score.penup()
        self.score.hideturtle()
        self.score.goto(0, 250)
        self.score.write("Hit: {}   Missed: {}".format(self.hit, self.miss), align='center', font=('Courier', 24, 'normal'))

        # -------------------- Keyboard control ----------------------

        self.win.listen()
        self.win.onkey(self.sumobot_right, 'Right')
        self.win.onkey(self.sumobot_left, 'Left')
        self.win.onkey(self.sumobot_up, 'Up')
        self.win.onkey(self.sumobot_down, 'Down')

    # Sumobot movement

    def sumobot_right(self):

        x = self.sumobot.xcor()
        self.sumobot.setx(x+20)

    def sumobot_left(self):

        x = self.sumobot.xcor()
        self.sumobot.setx(x-20)
            
    def sumobot_up(self):

        y = self.sumobot.ycor()
        self.sumobot.sety(y+20)
    
    def sumobot_down(self):

        y = self.sumobot.ycor()
        self.sumobot.sety(y-20)

    def run_frame(self):

        self.win.update()

        # Sumobot moving

        self.sumobot.setx(self.sumobot.xcor() + self.sumobot.dx)
        self.sumobot.sety(self.sumobot.ycor() + self.sumobot.dy)

        # Enemy and Sumobot collision

        # Enemy Arena contact

        if sqrt(pow(self.sumobot.xcor(), 2) + pow(self.sumobot.ycor(), 2)) > 380:
            self.sumobot.goto(0, -380)
            self.miss += 1
            self.score.clear()
            self.score.write("Hit: {}   Missed: {}".format(self.hit, self.miss), align='center', font=('Courier', 24, 'normal'))
            self.reward -= 3
            self.done = True
        else:
            self.reward +=3

#         # Enemy Sumobot collision

#         if abs(self.enemy.ycor() + 250) < 2 and abs(self.sumobot.xcor() - self.enemy.xcor()) < 55:
#             self.enemy.dy *= -1
#             self.hit += 1
#             self.score.clear()
#             self.score.write("Hit: {}   Missed: {}".format(self.hit, self.miss), align='center', font=('Courier', 24, 'normal'))
#             self.reward += 3

    # ------------------------ AI control ------------------------

    # 0 do nothing
    # 1 move left
    # 2 move right
    # 3 move up
    # 4 move down

    def reset(self):

        self.sumobot.goto(0, -380)
#         self.enemy.goto(0, 100)
        return [self.sumobot.xcor()*0.01, self.sumobot.ycor()*0.01, self.sumobot.dx, self.sumobot.dy]

    def step(self, action):

        self.reward = 0
        self.done = 0

        if action == 0:
            self.reward += .01

        if action == 1:
            self.sumobot_left()
            self.reward += .1
            
        if action == 2:
            self.sumobot_right()
            self.reward += .1
            
        if action == 3:
            self.sumobot_up()
            self.reward += .1
            
        if action == 4:
            self.sumobot_down()
            self.reward += .1

        self.run_frame()

        state = [self.sumobot.xcor()*0.01, self.sumobot.ycor()*0.01, self.sumobot.dx, self.sumobot.dy]
        return self.reward, state, self.done


# ------------------------ Human control ------------------------
#
env = Sumobot()

while True:
     env.run_frame()
```
